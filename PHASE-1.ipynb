CREATE SERIES 

```python
import pandas as pd 
sales =  pd.Series([2,3,5])
print(sales)
```

INDEXING AND SLICING  

```python
print(sales[1])
print(sales[0:2])
```

ATTRIBUTES 

```python
print(sales.index)
```

```python
print(sales.values)
print(sales.head)
print(sales.tail(1))
print(sales.sort_values)
print(sales.count())
```

```python
print(sales.isnull())
```

```python
sales_missing  = pd.Series([2,None,5])
print(sales_missing)
print(sales_missing.fillna(12))
```

```python
print(sales_missing)
sales_missing = sales_missing.fillna(12)
print(sales_missing)
```

# PANDAS DATAFRAME

```python
import pandas as pd
data = pd.DataFrame({
    'Name':['ajay','vijay'],
    'lemons':[2,3]
})
print(data)
```

```python
s1 = pd.Series(data=[1,2,3,4,5])
s2 = pd.Series(data=[10,20,30,40,50])
data = pd.DataFrame({'c1':s1,'c2':s2 })
print(data)
print(s1.shape)
print(data.shape)
print(data.columns)
```

#ACCESSING DATA :  difference between locc and iloc 

```python
print(data.loc[0,'c1'])
print(data.iloc[1,1])
print(data.mean())
```
